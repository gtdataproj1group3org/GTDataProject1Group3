```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

nfl_data_to_load =  pd.read_csv("../data/mike.csv")
n_games_df = pd.DataFrame(nfl_data_to_load)



```

```python
def homepass (row):
    if row['home_team'] == row['posteam'] and int(row['pass_attempt']) == 1:
        return 1
    return 0
def awaypass (row):
    if row['away_team'] == row['posteam'] and int(row['pass_attempt']) == 1:
        return 1
    return 0
def homerush (row):
    if row['home_team'] == row['posteam'] and int(row['rush_attempt']) == 1:
        return 1
    return 0
def awayrush (row):
    if row['away_team'] == row['posteam'] and int(row['rush_attempt']) == 1:
        return 1
    return 0
n_games_df['homepass'] = n_games_df.apply (lambda row: homepass(row), axis=1) 
n_games_df['homerush'] = n_games_df.apply (lambda row: homerush(row), axis=1) 
n_games_df['awaypass'] = n_games_df.apply (lambda row: awaypass(row), axis=1) 
n_games_df['awayrush'] = n_games_df.apply (lambda row: awayrush(row), axis=1) 
n_games_df
```

```python
setitupz = n_games_df.groupby(['game_id', 'home_team','away_team'])
#setitupz['homepass'].count()
#setitupz['awaypass'].sum()

#setitupz()

#setitupz['total_home_score'].max()
#setitupz['total_away_score'].max()
#setitupz['pass_attempt'].sum()
#setitupz['rush_attempt'].sum()

df_final=pd.DataFrame({'home_score':setitupz['total_home_score'].max(),'away_score':setitupz['total_away_score'].max(),'home pass attempts':setitupz['homepass'].sum(),'home rush attempts':setitupz['homerush'].sum(),'away rush attempts':setitupz['awayrush'].sum(),'away pass attempts':setitupz['awaypass'].sum()})
df_final
```

```python
#lambda needs index
df_final = df_final.reset_index()
df_final
```

```python
def winner (row):
    if row['home_score'] > row['away_score']:
        return row['home_team']
    elif row['home_score'] < row['away_score']:
        return row['away_team']
    return 'TIE'
df_final['winner'] = df_final.apply (lambda row: winner(row), axis=1) 
df_final
```

```python
def wrushpercentage (row):
    if row['winner'] == row['home_team']:
        return  row['home rush attempts'] * 100/(row['home rush attempts'] + row['home pass attempts'])
#        return rushp
    elif row['winner'] == row['away_team']:
        return  row['away rush attempts'] * 100/(row['away rush attempts'] + row['away pass attempts'])
#        return rushp
    return 0 
df_final['wrushpercentage'] = df_final.apply (lambda row: wrushpercentage(row), axis=1) 
```

```python
def lrushpercentage (row):
    if row['winner'] != row['home_team']:
        return  row['home rush attempts'] * 100/(row['home rush attempts'] + row['home pass attempts'])
#        return rushp
    elif row['winner'] != row['away_team']:
        return  row['away rush attempts'] * 100/(row['away rush attempts'] + row['away pass attempts'])
#        return rushp
    return 0 
df_final['lrushpercentage'] = df_final.apply (lambda row: lrushpercentage(row), axis=1) 
```
