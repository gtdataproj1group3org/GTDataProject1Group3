```python
# import dependencies
import pandas as pd
import numpy as np

# define string where data resides
str_PlayData = "../data/NFL-PlayByPlay-2009-2018.csv"
str_PlayDataReduced_Output = "../data/SAMPLES-NFL-PlayByPlay-2009-2018.csv"
```

```python
# Read in play by play data
df_PlaysRawData = pd.read_csv(str_PlayData)
```

```python
# check to see we got data
df_PlaysRawData.head()
```

```python
# reduce dataset down by 90% by sampling based on game_date
list_GameDays = df_PlaysRawData["game_date"].unique()

# setup list of indices to grab samples
list_indices = np.arange(0,len(list_GameDays),10)

# reduce list of game days based on indices
list_GameDays_reduced = [list_GameDays[index] for index in list_indices]

# reduce original dataframe based on samples of game days
df_PlaysReduced = df_PlaysRawData[df_PlaysRawData["game_date"].isin(list_GameDays_reduced)]

# validate the reduced value counts
df_PlaysReduced.count()
```
